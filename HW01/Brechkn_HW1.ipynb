```python
from sklearn.linear_model import LinearRegression

df_male['age'].fillna(df_male['age'].mean(), inplace=True)
df_male['age_o'].fillna(df_male['age_o'].mean(), inplace=True)

X = df_male['age'].values.reshape(-1, 1)
y = df_male['age_o'].values

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

plt.figure(figsize=(10, 6))
plt.scatter(df_male['age'], df_male['age_o'], alpha=0.6)
plt.plot(df_male['age'], y_pred, color='red', linewidth=2)
plt.title('Линейная регрессия: возраст vs ожидаемый возраст')
plt.xlabel('Возраст (age)')
plt.ylabel('Ожидаемый возраст (age_o)')
plt.grid()
plt.show()
```

```python
df_female['age'].fillna(df_female['age'].mean(), inplace=True)
df_female['age_o'].fillna(df_female['age_o'].mean(), inplace=True)

X = df_female['age'].values.reshape(-1, 1)
y = df_female['age_o'].values

model1 = LinearRegression()
model1.fit(X, y)

y_pred = model1.predict(X)

plt.figure(figsize=(10, 6))
plt.scatter(df_female['age'], df_female['age_o'], alpha=0.6)
plt.plot(df_female['age'], y_pred, color='red', linewidth=2)
plt.title('Линейная регрессия: возраст vs ожидаемый возраст')
plt.xlabel('Возраст (age)')
plt.ylabel('Ожидаемый возраст (age_o)')
plt.grid()
plt.show()
```

```python
import scipy.stats as stats

stats.probplot(df_male['age'], dist="norm", plot=plt)
plt.title('Q-Q график для возраста')
plt.grid()
plt.show()
```

```python
stat, p_value = stats.shapiro(df_male['age'])
print(f'Статистика: {stat}, p-value: {p_value}')
if p_value > 0.05:
    print('Данные имеют нормальное распределение')
else:
    print('Данные не имеют нормальное распределение')
```
