```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv("Life_train.csv")
df.shape
```

```python
df.isnull().sum()
```

```python
np.isinf(df).sum()
```

```python
df.fillna(df.median(), inplace=True)

X = df.drop(columns=['Life expectancy ']) 
Y = df['Life expectancy ']

X
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=1/3, random_state=111)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)

model = LinearRegression()
model.fit(X_train, Y_train)

y_pred_val = model.predict(X_val)

mse = mean_squared_error(y_pred_val, Y_val)
print(f"MSE: {mse}")
```

```python
X_full = scaler.fit_transform(X)
model.fit(X_full, Y)
```

```python
df_test = pd.read_csv("Life_test.csv")
print(df_test.isnull().sum())
print(np.isinf(df).sum())
```

```python
df_test.fillna(df_test.median(), inplace=True)

X_test = df_test.iloc[:, :].values

X_test = scaler.transform(X_test)

Y_pred_test = model.predict(X_test)

df_ans = pd.DataFrame({'Life expectancy': Y_pred_test})
df_ans['index'] = df_ans.index
df_ans = df_ans[['index', 'Life expectancy']]
df_ans.to_csv('life_expectancy1.csv', index=False)
```
