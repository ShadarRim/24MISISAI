```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Physical_Activity_Monitoring_unlabeled.csv')
```

```python
plt.figure(figsize=(30, 25))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.show()
```

```python
columns_to_drop = ['timestamp', 'subject_id', 'handAcc16_1', 'handAcc16_2', 'handAcc16_3', 'chestAcc16_1', 'chestAcc16_2', 'chestAcc16_3', 'ankleAcc16_1', 'ankleAcc16_2', 'chestOrientation3']
df = df.drop(columns=columns_to_drop)
```

```python
plt.figure(figsize=(30, 25))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.show()
```

```python
print(df.isna().sum())
```

```python
df = df.fillna(df.median())
```

```python
print(df.isna().sum())
```

```python
df_standard_scaled = StandardScaler().fit_transform(df)

df_scaled = pd.DataFrame(df_standard_scaled, index=df.index, columns=df.columns)
```

```python
df_scaled
```

```python
from sklearn.decomposition import PCA

pca = PCA().fit(df_scaled)

plt.figure(figsize=(10,7))
plt.plot(np.cumsum(pca.explained_variance_ratio_), color='k', lw=2)
plt.xlabel('components')
plt.ylabel('explained variance')
plt.xlim(0, 63)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.show();
```

```python
pca = PCA(n_components=26)

pComponents = pca.fit_transform(df_scaled)

pcaDf = pd.DataFrame(data=pComponents)

evr = pca.explained_variance_ratio_
evr.sum()
```

```python
from sklearn.cluster import KMeans

random_state = 42
n_clusters = 7

clusterer = KMeans(n_clusters=n_clusters, random_state = random_state)

cluster_labels = clusterer.fit_predict(pComponents)

inertia = clusterer.inertia_
```

```python

print(f"{n_clusters} - {inertia}")


# 4 - 15033604.95311292
# 5 - 14456327.757419879
# 6 - 13924226.032917652
# 7 - 13690927.026064638
# 8 - 12911770.155933475
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

inertia = []
silhouette_coefficients = []

for k in range(3, 15):
    kmeans = KMeans(n_clusters=k, random_state=1).fit(pComponents)
    inertia.append(np.sqrt(kmeans.inertia_))

plt.plot(range(3, 15), inertia, marker='s');
```

```python
predictions = pd.DataFrame(cluster_labels, columns=['activityID'])
predictions = predictions.reset_index()

predictions['activityID'].unique()
```

```python
transformation_rule = {5: 1, 3: 2, 2: 3, 0: 4, 6: 5, 4: 6, 1: 7}

predictions["activityID"] = predictions["activityID"].map(transformation_rule).astype(int)
predictions
```
