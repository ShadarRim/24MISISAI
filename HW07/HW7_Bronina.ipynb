```python
df = df.drop(columns=['timestamp', 'subject_id'])
```

### Заполнение пропусков

```python
df = df.interpolate()
df.isna().sum()
```

### Стандартизация данных

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
std_df = pd.DataFrame(scaler.fit_transform(df), index=df.index, columns=df.columns)
std_df
```

### Понижение размерности в данных (упрощение большого набора данных до меньшего набора с помощью метода главных компонент, при этом сохраняя значимые закономерности и тенденции)

```python
from sklearn.decomposition import PCA

pca = PCA().fit(std_df)

plt.figure(figsize=(10, 10))
plt.plot(np.cumsum(pca.explained_variance_ratio_), lw=2)
plt.xlabel('Components')
plt.ylabel('Explained variance')
plt.xlim(0, 52)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.axvline(26, c='b')
plt.axhline(0.92, c='r')

plt.show();
```

```python
pca = PCA(n_components=26)

new_components = pca.fit_transform(std_df)

new_df = pd.DataFrame(data = new_components)

pca.explained_variance_ratio_.sum()
```

```python
plt.figure(figsize=(40, 40))
sns.heatmap(new_df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
std_df.columns
```

```python
from sklearn.cluster import KMeans

inertia_total = []

for k in range(3, 10):
    kmeans = KMeans(n_clusters=k, random_state=1).fit(std_df)
    inertia_total.append(np.sqrt(kmeans.inertia_))

plt.plot(range(3, 10), inertia_total, marker='s');
plt.xlabel('$k$')
plt.ylabel('$J(C_k)$');
```

По общему датафрейму не видно явных перегибов, поэтому можем разделить данные на 3 части (hand, chest, ankle), чтобы определить оптимальное количество кластеров

```python
std_df_hand = std_df[['handTemperature', 'handAcc16_1', 'handAcc16_2', 'handAcc16_3',
       'handAcc6_1', 'handAcc6_2', 'handAcc6_3', 'handGyro1', 'handGyro2',
       'handGyro3', 'handMagne1', 'handMagne2', 'handMagne3',
       'handOrientation1', 'handOrientation2', 'handOrientation3',
       'handOrientation4']]

std_df_chest = std_df[['chestTemperature', 'chestAcc16_1', 'chestAcc16_2',
       'chestAcc16_3', 'chestAcc6_1', 'chestAcc6_2', 'chestAcc6_3',
       'chestGyro1', 'chestGyro2', 'chestGyro3', 'chestMagne1', 'chestMagne2',
       'chestMagne3', 'chestOrientation1', 'chestOrientation2',
       'chestOrientation3', 'chestOrientation4']]

std_df_ankle = std_df[['ankleTemperature',
       'ankleAcc16_1', 'ankleAcc16_2', 'ankleAcc16_3', 'ankleAcc6_1',
       'ankleAcc6_2', 'ankleAcc6_3', 'ankleGyro1', 'ankleGyro2', 'ankleGyro3',
       'ankleMagne1', 'ankleMagne2', 'ankleMagne3', 'ankleOrientation1',
       'ankleOrientation2', 'ankleOrientation3', 'ankleOrientation4']]
```

```python

inertia_hand = []

for k in range(3, 10):
    kmeans = KMeans(n_clusters=k, random_state=1).fit(std_df_hand)
    inertia_hand.append(np.sqrt(kmeans.inertia_))

plt.plot(range(3, 10), inertia_hand, marker='s');
plt.xlabel('$k$')
plt.ylabel('$J(C_k)$');
```

```python
inertia_chest = []

for k in range(3, 10):
    kmeans = KMeans(n_clusters=k, random_state=1).fit(std_df_chest)
    inertia_chest.append(np.sqrt(kmeans.inertia_))

plt.plot(range(3, 10), inertia_chest, marker='s');
plt.xlabel('$k$')
plt.ylabel('$J(C_k)$');
```

```python
inertia_ankle = []

for k in range(3, 10):
    kmeans = KMeans(n_clusters=k, random_state=1).fit(std_df_ankle)
    inertia_ankle.append(np.sqrt(kmeans.inertia_))

plt.plot(range(3, 10), inertia_ankle, marker='s');
plt.xlabel('$k$')
plt.ylabel('$J(C_k)$');
```

### Кластеризация

```python
from sklearn.cluster import KMeans

max_iter = 500
random_state = 42
n_clusters = 5

clusterer = KMeans(init='k-means++', n_clusters=n_clusters, max_iter = max_iter, random_state = random_state)

cluster_labels = clusterer.fit_predict(new_df)
inertia = clusterer.inertia_
inertia
```

### Подготовка ответа

```python
predictions = pd.DataFrame(cluster_labels, columns=['activityID'])
predictions = predictions.reset_index()

predictions
```

```python
predictions['activityID'].unique()
```

```python
mapping = {
    2: 1,
    0: 2,
    1: 3,
    4: 4,
    3: 5
}

predictions['activityID'] = predictions['activityID'].map(mapping)
predictions.head()
```
