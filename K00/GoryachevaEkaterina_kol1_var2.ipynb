## Списки

Вам дан список. Поменяйте местами первую и вторую половину списка.

```python
a = [i for i in range(10)]
b = [i for i in range(5)]

na = len(a)//2
nb = len(b)//2

a = a[na:] + a[:na]
print(a)
b = b[nb:] + b[:nb]
print(b)
```

```python
# ваше решение
a = [i for i in range(10)]
a[1:5:2], a[-2: -5: -2] = a[-2: -5: -2], a[1:5:2]
a
```

## Генераторы

Вам дан список. Создайте новый список, в котором останутся только чётные элементы кратные трём.

```python
# ваше решение
a = [i for i in range(28)]
print(a)
a = [i for i in a if (i%2==0) and (i%3==0)]
a
```

Вам дан список. Для него и ивертированного по порядку списка выведите пары (x, y) в новом списке.

```python
# ваше решение
a = [i for i in range(30)]
n = len(a)

a = [(a[i], a[n-i-1]) for i in range(n)]
a
```

## Встроенные функции

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("garments_worker_productivity.csv")
df
```

```python
# Какие типы данных присутствуют в датасете?
df_type = df.info()
# строки числа числа с плавающей точкой
```

```python
# Есть ли в датасете пропущенные значения? Если да, то сколько и где?
df.isnull().sum()
```

```python
# Постройте график распределения переменной actual_productivity.
plt.figure(figsize=(5,5))
sns.displot(df['actual_productivity'], bins=20)
```

```python
# Постройте heatmap для матрицы корреляции числовых переменных в датасете.
df_corr = df.drop(columns=["date", "quarter", "department", "day"])

plt.figure(figsize=(12,10))
sns.heatmap(df_corr.corr(),
            xticklabels=df_corr.corr().columns,
            yticklabels=df_corr.corr().columns,
            cmap='RdYlGn',
            center=0,
            annot=True)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
# Постройте pairplot для всех переменных.
plt.figure(figsize=(10,8))
sns.pairplot(df_corr)
plt.show()
```

```python
# Постройте boxplot для всех переменных и выведете объекты которые являются выбросами (строки из датафрейма).
fig, axes = plt.subplots(11, 1, figsize=(5,50))

name = df_corr.columns
# print(name)
for i in range(len(name)):
    sns.boxplot(df_corr[name[i]], ax=axes[i], width=0.3)
    title_i = name[i]
    axes[i].set_title(title_i)
```
