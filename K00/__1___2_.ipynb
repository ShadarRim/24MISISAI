## Списки

Вам дан список. Поменяйте местами первую и вторую половину списка.

```python
arr = [i for i in range(8)]

middle = len(arr) // 2
first_part = arr[:middle]
second_part = arr[middle:]
arr = second_part + first_part

arr
```

Вам дан список. Поменяйте в нём местами блоки элементов: каждый нечётный со первого по пятый с начала с каждым чётный со второго по пятый с конца.

```python
a = [i for i in range(20)]
print(a)
a[2:6:2], a[-2:-5:-2] = a[-2:-5:-2], a[2:6:2]

a
```

## Генераторы

Вам дан список. Создайте новый список, в котором останутся только чётные элементы кратные трём.

```python
arr = [0, 1, 2, 3, 6, 8, 9, 11, 12, 14, 15, 18]
arr = [arr[i] for i in range(len(arr)) if arr[i] % 3 == 0 and i % 2 == 0]
arr
```

## NumPy

Дан массив 16х16
Создайте новый массив 4х4 , элементы которого будут суммой блоков 4х4 исходного массива

```python
import numpy as np
arr = np.array([[i+j for j in range(16)] for i in range(16)])
print(arr)
arr2 = np.array([[arr[i*4:i*4+4, j*4:j*4+4].sum() for j in range(arr.shape[1]//4)] for i in range(arr.shape[0]//4)])
arr2
```

Дан массив, вставьте между каждым значением сумму соседних элементов

```python
arr = np.array([i for i in range(1, 15, 2)])
print(arr)
arr2 = np.insert(arr, np.arange(1, len(arr)), arr[:-1] + arr[1:])
arr2
```

```python
plt.figure(figsize=(5,5))

sns.distplot(df['actual_productivity'], bins=30)
# Постройте график распределения переменной actual_productivity.
```

```python
# Постройте heatmap для матрицы корреляции числовых переменных в датасете.
# Оставляем только числовые столбцы
numeric_df = df.select_dtypes(include=['number'])

# Строим корреляционную матрицу на основе числовых данных
plt.figure(figsize=(12, 10), dpi=80)
sns.heatmap(numeric_df.corr(), xticklabels=numeric_df.corr().columns,
            yticklabels=numeric_df.corr().columns, cmap='RdYlGn', center=0, annot=True)

plt.title('Correlogram', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
# Постройте pairplot для всех переменных.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


table = numeric_df.corr()
columns = table.columns
selected_columns = set()
for col in table:
    i = 0
    while columns[i] != col:
        if table[col].iloc[i] >= 0.3:
            print((col, columns[i]), 'corr = ', table[col].iloc[i])
        i += 1

print("Выбранные столбцы для pairplot:", selected_columns)


```

```python

plt.figure(figsize=(10,8), dpi= 80)
sns.pairplot(df)
plt.show()
```

```python
fig, axes = plt.subplots(1, 11, figsize=(30,10))
# Постройте boxplot для всех переменных и выведете объекты которые являются выбросами (строки из датафрейма).
i = 0
for col in df.columns:
  if df[col].dtype.name in ('int64', 'float64'):
    boxplot = sns.boxplot(df[col], ax=axes[i], width=0.3)
    boxplot.set_xlabel(col)
    i += 1
```

```python

for col in df.columns:
  if df[col].dtype.name in ('int64', 'float64'):
    Q1_wip = df[col].quantile(0.25)
    Q3_wip = df[col].quantile(0.75)
    IQR_wip = Q3_wip - Q1_wip
    print(f'Выбросы столбца {col}')
    print(df[(df[col] <= Q1_wip - 1.5 * IQR_wip) | (df[col] >= Q3_wip + 1.5 * IQR_wip)])
```
