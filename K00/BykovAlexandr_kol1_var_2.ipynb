## Списки

Вам дан список. Поменяйте местами первую и вторую половину списка.

```python
def ch(l: list) -> list:
    s_l = l[0:int(len(l) / 2)]
    l[0:int(len(l) / 2)] = l[int(len(l) / 2):]
    l[int(len(l) / 2):] = s_l
    return l
print(ch(['2', '4', '5', '6']))
print(ch(['2', '4', '5']))
```

Вам дан список. Поменяйте в нём местами блоки элементов: каждый нечётный со первого по пятый с начала с каждым чётный со второго по пятый с конца.

```python
def ch_2(l: list) -> list:
    f_b = l[0:5:2]
    s_b = l[-2:-6:-2]

    for i in range(2):
        l[i * 2] = s_b[i]
        l[len(l) - 2 - i * 2] = f_b[i]

    return l
print(ch_2(['2', '3', '4', '5', '6', '7', '8', '9']))
```

## Генераторы

Вам дан список. Создайте новый список, в котором останутся только чётные элементы кратные трём.

```python
def n(l: list) -> list:
    new_l = [l[1 + i * 2] for i in range(len(l[1::2])) if l[1 + i * 2] % 3 == 0]
    return new_l
print(n([1, 2, 4, 6, 7, 8, 9, 12]))
```

Вам дан список. Для него и ивертированного по порядку списка выведите пары (x, y) в новом списке.

```python
def rev(l: list) -> list:
    rev_l = l[::-1]
    new_l = [(l[i], rev_l[i]) for i in range(len(l))]
    return new_l
print(rev([2, 4, 5, 6, 8]))
print(rev([2, 4, 5, 6, 8, 9]))
```

## Встроенные функции

Вам дана строка. Выведите сумму ascii-кодов элементов в строке.

```python
def ascii_sum(s: str) -> int:
    summ = 0
    for e in s:
        summ += ord(e)
    return summ
print(ascii_sum("a"))
print(ascii_sum("ab"))
```

Вам дана строка. Выведите строку, выведите пары (элемент строки, порядковый номер).

```python
def pair_s_print(s: str) -> None:
    print(s)
    print(list(enumerate(s)))
pair_s_print("abc")
```

Дан массив 16х16
Создайте новый массив 4х4 , элементы которого будут суммой блоков 4х4 исходного массива

```python
arr = np.random.randint(0, 2, (16, 16))

def res(arr: np.array) -> np.array:
    new_arr = arr.reshape(4,4,4,4).sum(axis=(3,1))
    return new_arr

print(arr)
print(res(arr))
```

Дан массив, вставьте между каждым значением сумму соседних элементов

```python
arr = np.arange(1, 6)

def ins(arr: np.array) -> np.array:
    sum_middle = arr[1:] + arr[:-1]
    new_arr = np.array(0)
    new_arr.resize(len(arr) * 2 - 1)
    new_arr[0::2] = arr
    new_arr[1::2] = sum_middle
    return new_arr
print(ins(arr))
```
